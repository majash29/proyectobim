```python
## 6.0 Histograma (Edades)
edad = dataset["Edad"]
intervalos = range(min(edad), max(edad)) #calculamos los extremos de los intervalos

plot.hist(x=edad, bins=intervalos, color='#900C3F', rwidth=0.8) # Construccion del histograma
plot.title('Histograma de Edades') # Añade un titulo
plot.xlabel('Edades') #Etiqueta a la serie del eje X
plot.ylabel('Frecuencia') #Etiqueta a la serie del eje Y
plot.xticks(intervalos) #Divisiones para el eje X

plot.show() #dibujamos el histograma
```
